```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib ipympl
```

```python
N = 20
x, y, z = uniform(N)
```

```python
_, ax = plt.subplots(subplot_kw={"projection": "3d"})
origins = np.zeros_like(x)
ax.quiver(origins, origins, origins, x, y, z, length=1)
plt.show()
```

```python
_, ax = plt.subplots()
x, y, z = uniform(500000)
v = np.transpose([x, y, z])
ax.hist(np.linalg.norm(v, axis=1), density=True, bins=200)
x = np.linspace(0.001, 1, 100)
ax.plot(x, 1 / np.sqrt(x) / 2)
plt.show()
```

```python
def export(x, y, z):
    N = x.size
    print(f"const vec4 n[{N}] = vec4[](")
    print(",\n".join(f"    vec4({a}, {b}, {c}, 0)" for a, b, c in zip(x, y, z)))
    print(");")
```

```python
x, y, z = uniform(64)
export(x, y, z)
```
